# **Iris Flower Classification - ML Project**

# Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Reading Data From Dataset

```python
col=['Sepal Length','Sepal Width','Petal Length','Petal Width','Class_Levels']
df=pd.read_csv('iris.data',names=col)
```

```python
df.head()
```

# Checking Data-type,Shape and Non-Null Values

```python
df.info(col)
```

# Data Visualization

```python
sns.pairplot(df,hue='Class_Levels')
```

# Separating Input and Output Columns

```python
data=df.values
X=data[:,0:4]
Y=data[:,4]
```

```python
print(X)
```

```python
print(Y)
```

# Splitting the Data Into Training and Testing

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

# Model 1:- Support Vector Machine Algorithm

```python
from sklearn.svm import SVC
model_svc=SVC()
model_svc.fit(X_train,Y_train)
```

```python
prediction1=model_svc.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction1)*100)
```

# Checking the Output's

```python
for i in range(len(Y_test)):
  print(Y_test[i],prediction1[i])
```

# Model 2:- Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
model_LR.fit(X_train,Y_train)
```

```python
prediction2=model_LR.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction2)*100)
```

# Checking the Output's

```python
for i in range(len(Y_test)):
  print(Y_test[i],prediction2[i])
```

# Model 3:- Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
model_DTC= DecisionTreeClassifier()
model_DTC.fit(X_train,Y_train)
```

```python
prediction3=model_DTC.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction3)*100)
```

# Checking the Output's

```python
for i in range(len(Y_test)):
  print(Y_test[i],prediction2[i])
```

# Detailed Classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,prediction1))
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,prediction2))
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,prediction3))
```
